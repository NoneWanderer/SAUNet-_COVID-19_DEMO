```python
import sys
import math
import random
import warnings
import numpy as np
import pandas as pd
from tqdm import tqdm
from skimage.io import imread, imshow, imread_collection, concatenate_images
from skimage.transform import resize
from sklearn.metrics import recall_score,f1_score
from keras import backend as K
import tensorflow as tf
import os
import warnings
from skimage.io import imread, imshow, imread_collection, concatenate_images
import matplotlib.pyplot as plt
from medpy import metric
```

```python
IMG_WIDTH = 256
IMG_HEIGHT = 256
IMG_CHANNELS = 1
TRUE_PATH = 'H:/TestResults/test_mask_org/'
PRED_PATH = 'H:/TestResults/unet++/'
```

```python
mask_gt = []
mask_pred = []
sys.stdout.flush()
file1 = os.listdir(TRUE_PATH)
file1.sort(key=lambda x:int(x[4:-4]))
file2 = os.listdir(PRED_PATH)
file2.sort(key=lambda x:int(x[7:-4]))
#print(file1)
#print(file2)
for i in tqdm(file1):
    img = imread(TRUE_PATH + i)
    img = resize(img, (IMG_HEIGHT, IMG_WIDTH),mode='constant', preserve_range=True)
    mask_gt.append(img)
for n in tqdm(file2):
    mask = imread(PRED_PATH + n)
    mask = resize(mask, (IMG_HEIGHT, IMG_WIDTH),mode='constant', preserve_range=True)
    mask_pred.append(mask)
```

```python
for i in range(0,1375,1):
    dice=metric.dc(mask_gt[i],mask_pred[i])
    print(dice)
```

```python
for i in range(0,1375,1):
    Hausdorff=metric.hd(mask_gt[i],mask_pred[i])
    print(Hausdorff)
```

```python
for i in range(0,1375,1):
    recall=metric.recall(mask_gt[i],mask_pred[i])
    print(recall)
```

```python
for i in range(0,1375,1):
    spe=metric.specificity(mask_gt[i],mask_pred[i])
    print(spe)
```

```python
for i in range(0,1375,1):
    IoU=metric.jc(mask_gt[i],mask_pred[i])
    print(IoU)
```
